```python
# Pour texte 2 (on peut faire une variante de la fonction compte_bigrammes pour les trigrammes .... n-grammes.)

def compte_bigrammes(texte_crypte):
    cpt_bigrammes = {}
    liste = []
    alphabet = "abcdefghijklmnopqrstuvwxyz"

    # Compte les bigrammes
    for i in range(len(texte_crypte) - 1):
        if texte_crypte[i] in alphabet and texte_crypte[i+1] in alphabet:
            # ne compte pas les bigrammes comprenants des caractères autres
            # que des lettres (de a à b).
            bigramme = texte_crypte[i] + texte_crypte[i + 1]
            cpt_bigrammes[bigramme] = cpt_bigrammes.get(bigramme, 0) + 1
            # Si bigramme dans dico ajoute 1, sinon renvoie 0 (rien)


    # Supprime les bigrammes qui n'apparaissent qu'une fois (car pas interessant pour notre étude.).
    for paire, valeur in cpt_bigrammes.items():
        if valeur != 1:
            liste.append([paire, valeur])

    # Range la liste dans ordre alphabétique
    liste = sorted(liste)

    return liste


texte = "gx qosvlnkd wkvlkxo xiu vscx qno yd fsu cx qniix cx unkggx kdvsddyx xu vsdukxdu g'kdckvx. gxi gxuuoxi cy fsu cx qniix qxofxuuxdu cx cxvngxo gxi gxuuoxi cy fxiinmx sokmkdng fscygs 26. ixygxi gxi gxuuoxi cx n n a isdu vlkwwoxxi."

print(compte_bigrammes(texte))
```
